```python
# Suraj wants to analyze the situation of covid-19 in India. He has a dataset of covid 19 cases in India state-wise. 
# The dataset contains nine columns:
# State/UTs	 
# Total Cases	 
# Active	
# Discharged	
# Deaths	 
# Active Ratio	 
# Discharge Ratio	 
# Death Ratio	 
# Population

# Help Suraj in finding the following things:
# Total number of cases in India.
# Find all the states where the death ratio is greater than 1 and the population is greater than 4000000.
# Show the count, mean, standard deviation, minimum values, and maximum values of all the columns ( Hint: Use the describe() method.)
# Find the data type of each column. (Hint: Use info() method )
```

```python

import pandas as pd
import numpy as np

data=pd.read_excel("csv_Data/Latest Covid-19 India Status.xlsx")
```

```python
df=pd.DataFrame(data)
```

```python
# Total number of cases in India.
df["Total Cases"].sum()
```

```python
# Find all the states where the death ratio is greater than 1 and the population is greater than 4000000.

dr=df["Death Ratio"]>1
pop=df["Population"]>4000000

df.loc[(dr&pop)]

# or
# df[(df["Death Ratio"]>1)&(df["Population"]>4000000)]
```

```python
# Show the count, mean, standard deviation, minimum values, and maximum values of all the columns ( Hint: Use the describe() method.)

df.describe()
```

```python
# Find the data type of each column. (Hint: Use info() method )

df.info()
```

```python
# Average active cases in India.


df.Active.sum()
```

```python
# State with lowest active cases.
# State with highest active cases.

df[df.Active==df.Active.min()]
```
