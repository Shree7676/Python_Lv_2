```python
import pandas as pd
import seaborn as sns  
import numpy as np
sns.set(color_codes=True) # adds a nice background to the graphs

auto = pd.DataFrame(np.random.randn(50,5))     #creating a dummy data frame
```

```python
# It can also be useful to visualize a relationship between two variables. The easiest way to do this in seaborn is to use the jointplot() function, which creates a scatterplot of the two variables along with the histograms of each next to the appropriate axes.
sns.jointplot(x=auto[2], y=auto[3])
# This will plot the data points of x vs y and also show the histogram of both x and y outside of the plot
# This is a great way to see if x and y show some kind of dependancy between each other
```

```python
# We can make a hex bin plot that breaks the 2D area into hexagons and the number of points in each hexagon determines the color
sns.jointplot(x=auto[2], y=auto[3], kind="hex")
# This is a great way to see where most of your data is concerntrated at.
# The darker the color, the more points are present in the hexagon
```

```python
# We can make a 2D estimation of the density also

sns.jointplot(x=auto[2], y=auto[3], kind="kde")

# This will plot a smooth contour graph, 
# This is a great way to see where your data is concentrated
```
