# ADDING 2 SERIES

```python
import pandas as pd
import numpy as np

fil=pd.read_excel("csv_Data/auto-mpg1.xlsx")
df=pd.DataFrame(fil)
```

```python
# 1 miles == 1.609344 km
# 1.00pund == 0.45359kg

mpg_kmg=df.mpg*1.609344
wght_kg=df.weight*0.45359
```

# METHOD 1

```python
pd.merge(mpg_kmg,wght_kg,right_index=True,left_index=True)
```

# METHOD 2

```python
d={"mpg_kmg":mpg_kmg,
   "wght_kg":wght_kg}

pd.concat(d,axis=1)
```

```python
# To extend series
pd.DataFrame(mpg_kmg.append(wght_kg,ignore_index=True))
```
